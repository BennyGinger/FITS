```python
import logging
from time import sleep

from progress_bar.progress import ProgressManager
from progress_bar.logging import attach_buffer_handler, detach_handler


def main() -> None:
    # IMPORTANT: do NOT call basicConfig in real fits.
    # This is just a smoke test.
    root = logging.getLogger()
    root.setLevel(logging.DEBUG)

    pm = ProgressManager.create(max_log_lines=8)

    # attach buffer capture to root logger, refresh UI on each log
    h = attach_buffer_handler(logging.getLogger(), pm.log_buffer, level=logging.INFO, on_emit=pm.refresh)

    logger = logging.getLogger("fits.smoke")

    with pm:
        pm.set_experiment("exp_1")
        tid = pm.add_task("Processing exp_1", total=20)
        for i in range(20):
            if i % 5 == 0:
                logger.info("exp_1 step %s", i)
            pm.advance(tid)
            sleep(0.5)

        pm.set_experiment("exp_2")
        tid2 = pm.add_task("Processing exp_2", total=10)
        for i in range(10):
            if i % 3 == 0:
                logger.warning("exp_2 warn %s", i)
            pm.advance(tid2)
            sleep(0.7)

    detach_handler(logging.getLogger(), h)

main()
```
